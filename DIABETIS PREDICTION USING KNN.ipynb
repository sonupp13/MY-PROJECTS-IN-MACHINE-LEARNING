```python
import pandas as pd
```

```python
data= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/diabetes.csv')
data
```

```python
data_new= data
import numpy as np
```

```python
# there are 0 values in the dataset now we are reolacing it with null values (pregnancy is not considered in diabetis)
data[['Glucose','BloodPressure','SkinThickness','Insulin'	,'BMI']]=data_new[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
# Replacing NaN with mean values
data_new['Glucose'].fillna(data_new['Glucose'].mean(),inplace=True)
data_new['BloodPressure'].fillna(data_new['BloodPressure'].mean(),inplace=True)
data_new['BMI'].fillna(data_new['BMI'].mean(),inplace=True)
data_new['Insulin'].fillna(data_new['Insulin'].mean(),inplace=True)
data_new['SkinThickness'].fillna(data_new['SkinThickness'].mean(),inplace=True)
```

```python
# Feature scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler(feature_range=(0,1)) # already ulla fn ne sc lekk call cheythu and (o ,1 ) means motham values 0 and 1 nte ullil set cheythu to get mean 0 and std 1
data_scaled=sc.fit_transform(data_new) # full data set converted to make mean 0 and s.d 1
```

```python
data_scaled= pd.DataFrame(data_scaled)
data_scaled
```

```python
# Selecting features - [Glucose, Insulin, BMI, Age]
x = data_scaled.iloc[:, [1, 4, 5, 7]].values  # full rows are selectd using ' : ' and [1, 4, 5, 7] these coluns are selected Glucose, Insulin, BMI, Age
y = data_scaled.iloc[:, 8].values # only output colunmn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=25,metric='minkowski',p=2)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score, precision_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
```
